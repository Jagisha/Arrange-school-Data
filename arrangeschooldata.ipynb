```python
ndf=primary_mix.merge(middle_mix,on=['STATE/U.T.'])
alldata=ndf.merge(high_school_mix,on=['STATE/U.T.'])
alldata.shape
```

```python
combined_data=alldata.iloc[-1,:]
combined_data
```

```python
combined_data.index
```

```python
combined_data.index=[col.strip() for col in combined_data.index] #comprehensive technique
combined_data.index
```

```python
combined_data.rename(index={'enrolment in class IX  - Girls': 'enrolment in class IX - Girls '},inplace=True)
```

```python
combined_data.index
```

```python
enrolment=pd.Dataframe(columns=['Boys' , 'Girls' , 'total'])
classes=['I' , 'II' , 'III' , 'IV', 'V', 'VI', 'VII' , 'VIII' , 'IX' , 'X']
for c in classes:
    boys= combined_data['enrolment in class ' + c + ' - boys']
    girls= combined_data['enrolment in class ' + c + ' - girls']
    total= combined_data['enrolment in class ' + c +' - total']
    enrolment= enrolment.append({'boys':boys, 'Girls':girls,'Total':total},ignore_index=True)
```
